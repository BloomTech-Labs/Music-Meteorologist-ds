```python
import pandas as pd
import sqlite3
import pickle
import numpy as np
```

```python
con = sqlite3.connect("song_list12.db")
df = pd.read_sql_query("SELECT * from songs", con)
con.close()
df = df.drop(['level_0'], axis=1)
```

```python
df.head(1)
```

```python
workdf = df
workdf.shape
```

```python
is_available =  workdf['availability']==1.0
workdf_available = workdf[is_available]
workdf_available.shape
```

```python
workdf_available.head()
```

```python
workdf_available = workdf_available.drop(['genres_stripped', 'availability', 'index'], axis=1)
```

```python
workdf_available.head()
```

```python
workdf_available.reset_index(inplace=True)
```

```python
workdf_available.head()
```

```python
pickle.dump( workdf_available, open( "song_list_v6.pkl", "wb" ) )
```
